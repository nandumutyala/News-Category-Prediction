```python
# Splitting the dataset into Training and Test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_final.index.values,df_final.label.values,test_size=0.20,random_state=42,stratify=df_final.label.values)
```

```python
df_final['data_type'] = ['not_set']*df_final.shape[0]
```

```python
df_final
```

```python
df_final.loc[X_train, 'data_type'] = 'train'
df_final.loc[X_test, 'data_type'] = 'test'
```

```python
#Text vectorization using TF-IDF
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_idf = TfidfVectorizer()
#applying tf idf to training data
X_train_tf = tf_idf.fit_transform(df_final[df_final['data_type']=='train']['clean_text'])
```

```python
X_train_tf.shape
```

```python
X_test_tf = tf_idf.transform(df_final[df_final['data_type']=='test']['clean_text'])
```

```python
X_test_tf.shape
```

```python
#Applying classifier
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
#Intialising Classifier
clf = LogisticRegression(random_state=0)
```

```python
#Training
clf.fit(X_train_tf, y_train)
```

```python

# Prediction
y_pred = clf.predict(X_test_tf)

```

```python
#Evaluation metrics
acc = accuracy_score(y_test, y_pred)
f1_score1 = f1_score(y_test, y_pred,average='weighted')
print("Logistic Regression model accuracy (in %):", acc*100)
print('Logistic Regression model F1 score :',f1_score1)
```
